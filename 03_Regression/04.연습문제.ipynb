# 당뇨병 회귀 예측 비교 
### (다중선형회귀, 결정트리회귀)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df = pd.DataFrame(diabetes.data, columns = diabetes.feature_names) 
df['target'] = diabetes.target
df.head()
```

```python
X = diabetes.data
y = df.target
X.shape, y.shape
```

```python
# train / test데이터 셋 분리

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size = 0.1, random_state = 2021
) 
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## - 다중선형회귀

```python
#다중선형회귀
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train) # 학습
pred_lr = lr.predict(X_test) # 예측
```

```python
# 다중선형 회귀식 설명 #
Weight, bias = lr.coef_, lr.intercept_
rss= lr._residues

pred_lr[1], np.dot(Weight.T, X_test[1]) + bias
# pred[1] 은 즉, np.dot(Weight.T, X_test[1]) + bias
```

```python
type(pred_lr)
```

## - 결정트리 회귀기

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(X_train,y_train) #학습
pred_dtr = dtr.predict(X_test) #예측
```

```python
#회귀식 설명# 선형 회귀식이랑 수식이 다름!을 확인함
Weight, bias = lr.coef_, lr.intercept_
rss= lr._residues
pred_dtr[0], np.dot(Weight.T, X_test[0]) + bias #다르네!
```

```python
type(pred_dtr)
```

```python
pred_dtr.shape
```

## 데이터 프레임 만들어 비교

```python
# 합칠 데이터 프레임1 
X_test_df = pd.DataFrame(X_test)
X_test_df.shape
```

```python
# 합칠 데이터 프레임2 
result = pd.DataFrame({ 
    'y_test' :y_test,
    '다중선형회귀 예측값': pred_lr,
    '결정트리 예측값':  pred_dtr,
})
result.reset_index(inplace =True) #인덱스 초기화 # 데이터 프레임 합칠때 실index기준으로 합치기위해
result.head()
```

```python
# 데이터셋 합치기
pd.concat([X_test_df, result], axis=1 )
```
